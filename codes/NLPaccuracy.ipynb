## load names from NC data 

```python
#data frame we will use to test the ACCURACY of nlp models
f = 'firstname'
l = 'lastname'
ncnamesdf = pd.DataFrame()


new = []
for n in list(ncvoterinfo['""first_name""']):
    n = n.replace('""', '')
    new.append(n)
del new[0:6]
new2 = []
for g in list(ncvoterinfo['""last_name""']):
    g = g.replace('""', '')
    new2.append(g)
del new2[0:6]
new3 = []
for s in list(ncvoterinfo['""race_code""']):
    s = s.replace('""', '')
    new3.append(s)
del new3[0:6]
    
ncnamesdf[f] = new
ncnamesdf[l]= new2
ncnamesdf['race_code'] = new3


#Goal
#names| P1| P2| Pn|Pmodels: P1*W1 + P2*W2 + Pn*Wn| true_race

racy = list(ncnamesdf['race_code'])



booli = []
for r in racy:
    if r == 'W':
        booli.append(0)
    
    else: 
        booli.append(1)
ncnamesdf['true_race'] = booli

# lastnames = []
# for l in lnamey:
#     if l == 'A':
#         lastnames.append(None)
#     else:
#         lastnames.append(l)

        
#Race Codes        
# race               description
# *******************************************************************************
# A                  ASIAN
# B                  BLACK or AFRICAN AMERICAN
# I                  AMERICAN INDIAN or ALASKA NATIVE
# M                  TWO or MORE RACES
# O                  OTHER
# P                  NATIVE HAWAIIAN or PACIFIC ISLANDER
# U                  UNDESIGNATED
# W                  WHITE
# *************************************************************************** /

# / ***************************************************************************
# Ethnic codes
# ethnicity          description
# *******************************************************************************
# HL                 HISPANIC or LATINO
# NL                 NOT HISPANIC or NOT LATINO
# UN                 UNDESIGNATED
# *************************************************************************** /
ncnamesdf #has race code and true race 
names = pd.DataFrame()
names['firstname'] = ncnamesdf[f]
names['lastname']=ncnamesdf[l]
names
```

## NLP #1: pred_wiki_name (this takes both the first and last names)

```python
from ethnicolr import pred_wiki_ln, pred_wiki_name

  
white = ['GreaterEuropean,WestEuropean,Italian_mean', 'GreaterEuropean,WestEuropean,Nordic_mean', 
          'GreaterEuropean,WestEuropean,Germanic_mean', 'GreaterEuropean,WestEuropean,French_mean',
          'GreaterEuropean,Jewish_mean', 'GreaterEuropean,British_mean','GreaterEuropean,EastEuropean_mean']

lastnames = list(ncnamesdf['lastname'])
firstnames = list(ncnamesdf['firstname'])

f = 'firstname'
l = 'lastname'



wikifulldf = pred_wiki_name(names,l, f,conf_int=0.9)
wikifulldf
a = pd.DataFrame(names, columns=[f, l])
a['white'] = wikifulldf[white].sum(axis=1)
a['not_white'] = 1- a['white']
                       

race = []
for n in list(a['not_white']):
    if .50 < n < 1.00:
        #print(True)
        race.append(1)
    else:
        #print(False)  
        race.append(0)

a['pred_race'] = race
a

#save as ethnicolor_wiki_full_predictions.csv
#null values:
#defalut is 1.00 not_white and for thre race column thats automatcially gunna be 0
```

## NLP #2: pred_fl_reg_name(df, 'last', 'first', conf_int=0.9) w fullname

```python
from ethnicolr import pred_fl_reg_name, pred_fl_reg_name_five_cat
odf = pred_fl_reg_name(names, l, f, conf_int=0.9)
b = pd.DataFrame(names, columns=[f, l])
b['white'] = odf['nh_white_mean']
b['not_white'] = 1 - b['white']


race = []
for n in list(b['not_white']):
    if n > .50:
        #print(True)
        race.append(1)
    else:
        #print(False)  
        race.append(0)
b['pred_race'] = race
b
```
