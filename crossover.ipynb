```python
import pandas as pd

asiandf = pd.read_csv("data/listings/yelp/yelp_asian_data.csv")
blackdf = pd.read_csv("data/listings/yelp/yelp_black_directory.csv")
latinxdf = pd.read_csv("data/listings/yelp/yelp_latinx_directory.csv")
nadf = pd.read_csv("data/listings/yelp/yelp_nativeamerican_data.csv")

asiandf


```

```python
#renaming colums to make merge easier 

blackdf.rename(columns={'Name': 'Business Name'}, inplace=True)
latinxdf.rename(columns={'Name': 'Business Name'}, inplace=True)
blackdf
```

```python
latinxdf
```

```python
#merge black and asian yelp data
yelp_black_asiandf = pd.merge(asiandf, blackdf, on= ["Business Name", "URL"], how="outer")
#new 
yelp_black_asiandf
```

```python
#merge latinx and native amaerican yelp data
yelp_na_latinxdf = pd.merge(latinxdf, nadf, on= ["Business Name", "URL"], how="outer")
#new
yelp_na_latinxdf 
```

```python
#merge into one minority yelp
yelp_minority_df= pd.merge(yelp_na_latinxdf, yelp_black_asiandf, on= ["Business Name", "URL"], how="outer")

yelp_minority_df
```

```python
minorityorigdf = pd.read_csv("data/listings/yelp/yelp_minority_directory.csv")
minorityorigdf.rename(columns={'Name': 'Business Name'}, inplace=True)
minorityorigdf
```

```python
yelp_allminority_df= pd.merge(yelp_minority_df, minorityorigdf, on= ["Business Name", "URL"], how="outer")
yelp_allminority_df
```

```python
yelp_allminority_df.to_csv('data/listings/yelp/yelpmerge.csv',index=False)
```
